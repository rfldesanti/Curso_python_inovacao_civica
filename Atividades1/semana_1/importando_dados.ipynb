## Aula 7: Carregando tabelas

É possível ler vários arquivos CSV e converter de uma só vez para dataframe Pandas. Para isso, além de usar as técnicas vistas nas aulas anteriores, importaremos a biblioteca glob.

#### A bilblioteca "glob" encontra todos os nomes de caminho(*pathnames*) que correspondem ao padrão informado. Essa biblioteca geralmente é utilizada para agrupar diversos arquivos em uma só variavel

```python
# importando a biblioteca glob
import glob
```

#### Pelo visto existem dois modos de trabalhar com a  função principal da biblioteca glob.
#### O primeiro modo é criando um iterador através do método "iglob". Esse método retorna uma lista de todos os caminhos que correspondem ao padrão informado, *porém sem armazená-los ao mesmo tempo* (ao contrário do que faz o método "glob", como veremos adiante).

#### Vamos verificar como funciona o método "iglob"

```python
# declarando uma lista para armazenar todos os caminhos
df = []
```

```python
# criando um loop for com o iterador retornado pelo método iglob para armaenzar na variável todos os caminhos
# encontrados pelo método
for dado in glob.iglob('base_dados/consulta_*.csv'):
    df_iglob  = pd.read_csv(dado, sep=';', encoding='latin_1')
    df.append(df_iglob)
df
```

```python
df_2 = []
for dado in glob.glob('base_dados/consulta_*.csv'):
    df_glob = pd.read_csv(dado, sep=';', encoding='latin_1')
    df_2.append(df_glob)
df_2
```

#### Já o método "glob" *armazena* todos os caminhos *pathnames* que correspondem ao padrão informado. No exemplo abaixo, podemos visualizar o que está armazenado na variável "dados_todos_estados". Com o método "iglob", essa visualização seria impossível

```python
# o método glob da biblioteca glob importa o caminho completo de todos os arquivos que correspondem ao padrão passado
# para ele. Portanto, a variável dados_todos_estados armazenará o caminho de todos os arquivos que estão em determinado
# local e possuem determinado nome
dados_todos_estados = glob.glob('base_dados/consulta_*.csv')
print(dados_todos_estados)
```

```python
df_RN = pd.read_csv(dados_todos_estados[0], sep=";", encoding="latin_1")
df_RN.shape
```

```python
df_RN = pd.read_csv('base_dados/consulta_cand_2020_RN.csv', sep=';', encoding='latin_1')
```

```python
df_PE = pd.read_csv(dados_todos_estados[1], sep=';', encoding='latin_1')
df_PE.shape
```

```python
# criamos uma variável para armazenar todos esses arquivos
lista_df = []
```

```python
# antes de armazenar os arquivos, porém, precisamos transformá-los em dataframes
# o loop for abaixo irá entrar na lista onde estão armanzeados os caminhos e transformar cada um dos arquivos
# que constam lá em dataframes. Após isso, irá colocar cada dataframe na variável criada anteriormente, a lista_df

for dados_estado in dados_todos_estados:
    df_estado = pd.read_csv(dados_estado, sep=';', encoding='latin_1')
    lista_df.append(df_estado)
lista_df
```

#### Vamos agora unir os dois dataframes. Faremos isso utilizando o método CONCAT

#### O método CONCAT concatena objetos panda *ao longo de um eixo* com lógica de conjunto opcional para os outros eixos

```python
# agora, queremos que todos os arquivos sejam agrupados(concatenados) em um arquivo só
# para isso, utiliza-se o método CONCAT, que cria um dataframe único, unindo todos os dataframes
# que estavam individualizados na variável lista_df
df_completo = pd.concat(lista_df, axis=0, ignore_index=True)
# sobre os parâmetros do método concat:
# axis=0: significa "concatenar as linhas (index)"; com "axis=1" significaria "concatenar colunas"
# ignore_index=True: pede para que o dataframe final, concatenado, não considere os índices dos dataframes originais
# com ignore_index=False, teríamos os índices dos dataframos originas no dataframe concatenado
df_completo.shape
```

### Como mostram as duas linhas acima, o número total de cidades se altera quando pesquisamos essa informação a partir de métodos diferentes. Ao filtrar os dados a partir dos Estados, temos um total de 351 cidades. Já quando analisamos o dataframe inteiro, sem filtrar os Estados inidividualmente, encontramos um total de 348 cidades. ### Para encontrar esses valores duplicados, vamos criar listas de valores únicos das cidades de cada Estado

```python
# criando uma lista de valores únicos para as cidades do Estado de RN. Vamos deixar a lista em ordem alfabética
ue_rn = total_cidades_rn['NM_UE'].sort_values().unique()
```

```python
# lista de valores únicos para as cidades do Estado de PE, também em ordem alfabética
ue_pe = total_cidades_pe['NM_UE'].sort_values().unique()
```

```python
# criando uma variável do tipo lista para inserir todas as cidades de cada estado
todas_cidades = []
```

```python
# criando um iterador for para adicionar cada uma das cidades de ue_rn na lista de todas_cidades
for cidade in ue_rn:
    todas_cidades.append(cidade)
```

```python
# o mesmo para as cidades de PE. Assim, a lista todas_cidades vai possuir as cidades de ambos os Estado
for cidade in ue_pe:
    todas_cidades.append(cidade)
```

```python
# conferindo o tamanho da lista todas_cidades
len(todas_cidades)
```

```python
# uma das formas de verificarmos quais são as cidades repetidas é criar uma variavel do tipo lista, que, a partir da
# ação de um iterador for, irá receber os valores que aparecerem mais de uma vez na lista todas_cidades;
# em outras palavras, o iterador irá contar, através do método count(), quantas vezes determinado valor aparece na lista.
# Caso o valor apareça mais de uma vez, ele será inserido na variável lista cidade[]; assim, saberemos quais nomes de cidades
# existem em ambos os Estados
cidade = []
for x in todas_cidades:
    if todas_cidades.count(x) > 1:
        cidade.append(x)
cidade
```

```python
# outra forma, mais simples, de encontrar os valores repetidos, é utilizar o método value_counts()
# o método value_counts(), porém, só existe EM DATAFRAMES OU SEMELHANTES, COMO SÉRIES, E NÃO EM LISTAS
# assim, para que possamos utilizar value_counts() em todas_cidades, temos, primeiro, que converter a variável do tipo
# lista em objeto Série. Assim, podemos criar uma variável para receber a Série todas_cidades e contar os valores
df_cidades = pd.Series(todas_cidades).value_counts()
df_cidades
# podemos ver como os valores duplicados (ou maiores que um, se fosse o caso) também podem ser encontrados de outra forma
```

```python
# podemos filtrar a série para verificar quais valores são maiores que 1 (ou seja, que existem em duplicidade na Série)
df_cidades[df_cidades > 1]
```
